```r
# ==================== SETUP ====================
options(repos = c(CRAN = "https://cloud.r-project.org"))

# Install all required packages at once
pkgs <- c("skimr", "dplyr", "visdat", "naniar", "mice", "ggplot2", "tidyr")
installed <- rownames(installed.packages())
to_install <- pkgs[!pkgs %in% installed]
if (length(to_install) > 0) install.packages(to_install, quiet = TRUE)

# Load all libraries
library(skimr)
library(dplyr)
library(visdat)
library(naniar)
library(mice)
library(ggplot2)
library(tidyr)
```

```r
# ==================== READ DATA ====================
data_path <- "/content/DataSet_No_Details.csv"
df <- read.csv("/content/DataSet_No_Details.csv")

str(df)
skim(df)
```

```r
# ==================== DATA PREPARATION ====================
cols_to_remove <- c("h_index_34", "h_index_56", "hormone10_1", "hormone10_2",
                    "an_index_23", "outcome", "factor_eth", "factor_h",
                    "factor_pcos", "factor_prl")

MD_df      <- df %>% select(-any_of(cols_to_remove))
factor_df  <- df %>% select(record_id, outcome, factor_eth, factor_h, factor_pcos, factor_prl)

str(MD_df)
summary(factor_df)
```

```r
# ==================== MISSING VALUES ====================
cat("Total NAs:", sum(is.na(MD_df)), "\n")
print(colSums(is.na(MD_df)))
skim(MD_df)

na_stats <- colMeans(is.na(MD_df)) * 100

cat("\n--- Columns with <= 35% missing ---\n")
na_stats_filtered <- na_stats[na_stats <= 35]
print(data.frame(Column = names(na_stats_filtered), NA_Percent = round(na_stats_filtered, 2), row.names = NULL))

cat("\n--- Columns with > 35% missing ---\n")
na_stats_filtered_1 <- na_stats[na_stats > 35]
print(data.frame(Column = names(na_stats_filtered_1), NA_Percent = round(na_stats_filtered_1, 2), row.names = NULL))
```

```r
# ==================== VISUALIZE MISSING DATA ====================
vis_miss(MD_df)        # Heatmap of NA patterns
gg_miss_var(MD_df)     # Barplot of missingness per variable
```

```r
# ==================== HANDLE MISSING DATA ====================
cols_to_remove1 <- c("hormone9", "hormone11", "hormone12", "hormone13", "hormone14")
handle_MD_df <- MD_df %>% select(-any_of(cols_to_remove1))
str(handle_MD_df)

# ==================== LITTLE'S MCAR TEST (Homework) ====================
# H₀: Data is MCAR    (fail to reject if p > 0.05)
# H₁: Data is not MCAR — MAR or MNAR  (reject if p <= 0.05)
```

```r
# ==================== IMPUTATION: RANDOM FOREST ====================
imputed_rf <- mice(handle_MD_df, method = "rf", m = 5, print = FALSE)
imputed_rf_final <- complete(imputed_rf)
```

```r
# Density plot: Original vs RF-imputed
ggplot(handle_MD_df, aes(x = hormone10_generated, fill = "Original")) +
  geom_density(alpha = 0.5) +
  geom_density(data = imputed_rf_final,
               aes(x = hormone10_generated, fill = "RF Imputed"), alpha = 0.5) +
  scale_x_continuous(limits = c(0, 2)) +
  scale_fill_manual(values = c("Original" = "steelblue", "RF Imputed" = "tomato")) +
  labs(title = "Density Plot: hormone10_generated — Original vs RF Imputed",
       fill = "Dataset") +
  theme_minimal()
```

```r
# ==================== IMPUTATION: PMM (Homework) ====================
imputed_pmm <- mice(handle_MD_df, method = "pmm", m = 5, print = FALSE)
imputed_pmm_final <- complete(imputed_pmm)   # FIX: was incorrectly using imputed_rf object
```

```r
# Density plot: Original vs PMM-imputed
ggplot(handle_MD_df, aes(x = hormone10_generated, fill = "Original")) +
  geom_density(alpha = 0.5) +
  geom_density(data = imputed_pmm_final,
               aes(x = hormone10_generated, fill = "PMM Imputed"), alpha = 0.5) +
  scale_x_continuous(limits = c(0, 2)) +
  scale_fill_manual(values = c("Original" = "steelblue", "PMM Imputed" = "seagreen")) +
  labs(title = "Density Plot: hormone10_generated — Original vs PMM Imputed",
       fill = "Dataset") +
  theme_minimal()
```

```r
# ==================== OUTLIER DETECTION ====================
# Lipids only
outliers_data <- imputed_rf_final %>%
  select(lipids1, lipids2, lipids3, lipids4, lipids5) %>%
  pivot_longer(everything(), names_to = "variable", values_to = "value")

ggplot(outliers_data, aes(x = variable, y = value)) +
  geom_boxplot(fill = "lightblue", alpha = 0.7) +
  labs(title = "Outlier Detection — Lipid Variables",
       x = "Variable", y = "Value") +
  theme_minimal()
```

```r
# All numeric variables
imputed_rf_final %>%
  select(where(is.numeric)) %>%
  pivot_longer(everything(), names_to = "name", values_to = "value") %>%
  ggplot(aes(y = value)) +
  geom_boxplot(fill = "lightblue", alpha = 0.7) +
  facet_wrap(~name, scales = "free") +
  labs(title = "Boxplots for Outlier Detection — All Numeric Variables") +
  theme_minimal()
```
